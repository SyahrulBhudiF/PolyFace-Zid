```python
import tensorflow as tf
import numpy as np
import os

BASE_DIR = os.getcwd()
MODEL_PATH = os.path.join(BASE_DIR, "models", "h5", "polyface_adagrad.h5")

print(f"Mencoba memuat model dari: {MODEL_PATH}\n")

try:
    model = tf.keras.models.load_model(MODEL_PATH)
    print("✅ Model berhasil dimuat.\n")

    target_layer = model.layers[-1]
    print(f"Menganalisis bobot dari layer: '{target_layer.name}' ({type(target_layer).__name__})\n")

    weights = target_layer.get_weights()

    if not weights:
        print("Layer ini tidak memiliki bobot untuk dianalisis (misalnya, InputLayer atau Dropout).")
    else:
        kernel_weights = weights[0]

        print("--- Statistik Bobot (Weights) ---")
        print(f"Bentuk (Shape) array bobot: {kernel_weights.shape}")
        print(f"Mean (rata-rata):   {np.mean(kernel_weights):.6f}")
        print(f"Std Dev (simpangan): {np.std(kernel_weights):.6f}")
        print(f"Min (nilai terkecil):    {np.min(kernel_weights):.6f}")
        print(f"Max (nilai terbesar):    {np.max(kernel_weights):.6f}")

except Exception as e:
    print(f"❌ Gagal memuat atau menganalisis model: {e}")
```
